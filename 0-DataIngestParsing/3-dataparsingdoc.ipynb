```python
from docx import Document as DocxDocument
import os
```

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
## Method 1 : Using DocxtxtLoader
print("Using Docx2txtloader")
try:
    docx_loader  = Docx2txtLoader("data/word/proposal.docx")
    docs = docx_loader.load()
    print(f" Loaded {len(docs)} document(s)")
    print(f" Content priview: {docs[0].page_content[:200]}...")
    print(f"Metadata: {docs[0].metadata}")
except Exception as e:
    print(f" Error: {e}")
```

```python
## Method 2: Using UnstructuredWordDocumentLoader
print("\nUsing UnstructuredWordDocumentLoader")
try:
    unstructured_loader = UnstructuredWordDocumentLoader("data/word/proposal.docx", mode = "elements")
    unstructured_doc = unstructured_loader.load()
    print(f" loaded {len(unstructured_doc)} document(s)")
    for i, doc in enumerate(unstructured_doc):
        print(f"\nElement {i+1}:")
        print(f" Type : {doc.metadata.get('category', 'unknown')}")
        print(f"Content: {doc.page_content[:100]}...")
except Exception as e:
    print(f" Error: {e}")
```
